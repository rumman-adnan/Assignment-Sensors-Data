```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
flower = sns.load_dataset("iris")

# Draw bar plot
sns.barplot(x="species", y="sepal_width", data=flower)
plt.title("Flowers Plot")

plt.show()



```

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
boat = sns.load_dataset("titanic")
boat
# Draw bar plot
sns.barplot(x="sex", y="alone", hue="who", data=boat, order=["female","male"], color="red",ci=None)
plt.title("Titanic Plot")

plt.show()



```
